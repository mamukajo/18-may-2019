# plot a line graph on **income vs year for the state of **california

```python
import pandas as pd
df=pd.read_csv('Untitled Folder/sample.csv')
y=df.values[4]
x=df.columns
x
y
```

```python
years=[]
for i in range(2,len(x)):
    years.append(x[i])
years

income=[]

for i in range(2,len(y)):
    
    income.append(y[i])
income
plt.xlabel('year')
plt.ylabel('income')
plt.title('income of california')

plt.plot(years,income)
plt.grid(True)
plt.show()

```

```python
width=1/2
plt.width=1/2
plt.title('Income of California')
plt.bar(years,income,width,color='Black')
plt.savefig('Untitled Folder/fig1.png')
plt.show()
```

# Problem-4

generate a scatter plot for 50 random numbers



```python
n=50
x=np.random.rand(n) #generate 50 randomnumbers

y=np.random.rand(n) #generate random numbers

colorlist=np.random.rand(n) # generate 

arealist=np.pi*(15*np.random.rand(50))**2


plt.scatter(x,y,s=arealist,c=colorlist,alpha=1)

plt.savefig('Untitled Folder/scatter.png')

```
